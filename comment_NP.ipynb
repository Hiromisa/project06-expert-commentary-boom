```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://newspicks.com/recommended'
```

```python
r = requests.get(url)
doc = BeautifulSoup(r.text, 'html.parser')
```

```python
print(doc.prettify())
```

```python
tests = doc.select(".recommended-category")
for test in tests:
    print("----------")
    print(test.select_one(".category-header").text.strip())
#     print(test.select_one(".recommended-picker-card").text.strip())
```

```python
experts = doc.select(".recommended-picker-card")
count = 0
for expert in experts:
    print("----------")
    print(expert.select_one(".user-name").text.strip())
    print(expert.select_one(".user-title").text.strip())
    print(expert.select_one("img")['src'])
    print(expert.select_one("a")['href'])
    count +=1
print(count)
```

```python
experts = doc.select(".recommended-picker-card")
rows = []
for expert in experts:
    row = {}
    row['name'] = expert.select_one(".user-name").text.strip()
    row['title'] = expert.select_one(".user-title").text.strip()
    row['photo'] = expert.select_one("img")['src']
    row['link'] = expert.select_one("a")['href']
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
```

```python
df['urls']= 'https://newspicks.com/' + df.link
```

```python
#create an empty list
desc_list = []
#loop through all urls
for url in df['urls'] :
    #send each url through beautiful soup
    r3 = requests.get(url)
    doc3 = BeautifulSoup(r3.text, "html.parser")
    #create an empty dictionary
    desc_dict={}
#     add description column
    desc_dict['likes'] = doc3.select_one(".likes").text.strip()
    desc_dict['profile'] = doc3.select_one(".profile").text.strip()
   #add urls columns for joining with the master data
    desc_dict["urls"] = url
    #add all dictionaries to the list
    desc_list.append(desc_dict)
```

```python
desc_list
```

```python
df2 = pd.DataFrame(desc_list)
```

```python
df2.head()
```

```python
master_df = df.merge(df2, on='urls')
```
