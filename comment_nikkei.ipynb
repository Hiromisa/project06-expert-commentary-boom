```python
experts = doc.select(".profile")
rows = []
for expert in experts:
    row = {}
    row['name'] = expert.select_one(".profile__name").text.strip()
    row['headline'] = expert.select_one(".profile__headline").text.strip()
    row['link'] = expert.select_one("a")['href']
    rows.append(row)
```

```python
rows
```

```python
df = pd.DataFrame(rows)
df.to_csv('expert.csv', index=False)
```

```python
df['urls']= 'https://www.nikkei.com' + df.link
```

```python
#create an empty list
desc_list = []
#loop through all urls
for url in df['urls'] :
    #send each url through beautiful soup
    r3 = requests.get(url)
    doc3 = BeautifulSoup(r3.text, "html.parser")
    #create an empty dictionary
    desc_dict={}
#     add description column
    desc_dict['description'] = doc3.select_one(".stream__header-description").text.strip()
    desc_dict['photo'] = doc3.select_one(".expert-comment-profile__thumbnail-image").get('src')
   #add urls columns for joining with the master data
    desc_dict["urls"] = url
    #add all dictionaries to the list
    desc_list.append(desc_dict)
```

```python
#put it in a dataframe
desc_df=pd.DataFrame(desc_list)
```

```python
master_df = df.merge(desc_df, on='urls').drop(columns='link')
```
